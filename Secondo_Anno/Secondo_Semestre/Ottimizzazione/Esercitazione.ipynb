```python
'''
2 variabili, una per ogni tipo di macchina, tanto una macchina deve passare per tutti gli stabilimenti
!!! METTERE 10 VARIABILI ERA SBAGLIATO, OGNI MACCHINA DEVE PASSARE PER TUTTI GLI STABILIMENTI

funzione obiettivo:
max z = 6000*xind + 2000*xagr

Vincolo B/C

oreRichiesteInd = [20, 10, 5, 15, 20]
oreRichiesteAgr = [12, 15, 10, 7, 9]

primi 3 tutti uguali posso fare for
oreRichiesteInd[0]*xind + oreRichiesteAgr[0]*xagr <= 180 linea A
oreRichiesteInd[1]*xind + oreRichiesteAgr[1]*xagr <= 135 linea B
oreRichiesteInd[2]*xind + oreRichiesteAgr[2]*xagr <= 200 linea C


oreRichiesteInd[0]*xind + oreRichiesteAgr[0]*xagr + oreRichiesteInd[3]*xind + oreRichiesteAgr[3]*xagr <= 300
oreRichiesteInd[2]*xind + oreRichiesteAgr[2]*xagr + oreRichiesteInd[4]*xind + oreRichiesteAgr[4]*xagr >= 100

Vincolo D/E

20*xind+10*xagr >= 25 (25% di 100)

Vincolo F

xagr >= xind/2

Vincolo G

xind + xagr > 4
xind >= 3
xagr >= 2
questi ultimi due vincoli li metto come lower bound
'''
```

```python
mod = gp.Model()
nstab = 5

#aggiungo 5 variabili per tipo di macchina
xind = mod.addVar(lb=3, vtype=GRB.INTEGER)
xagr = mod.addVar(lb=2, vtype=GRB.INTEGER)

#Vincolo B/C
oreRichiesteInd = [20, 10, 5, 15, 20]
oreRichiesteAgr = [12, 15, 10, 7, 9]
massimiPrime3Linee = [180, 135, 200]

#per le prime 3 linee i vincoli sono uguali
for i in range(3):
    mod.addConstr(oreRichiesteInd[i]*xind + oreRichiesteAgr[i]*xagr <= massimiPrime3Linee[i])

#ora i vincoli su linee combinate
mod.addConstr(oreRichiesteInd[0]*xind + oreRichiesteAgr[0]*xagr + oreRichiesteInd[3]*xind + oreRichiesteAgr[3]*xagr <= 300)
mod.addConstr(oreRichiesteInd[2]*xind + oreRichiesteAgr[2]*xagr + oreRichiesteInd[4]*xind + oreRichiesteAgr[4]*xagr >= 100)

#Vincolo D/E
mod.addConstr(20*xind+10*xagr >= 25)

#Vincolo F
mod.addConstr(xagr >= xind/2)

#Vincolo G
mod.addConstr(xind + xagr >= 4)
#gli altri due vincoli del punto G sono gia messi come lower bound

#Funzione obiettivo
mod.setObjective(6000*xind + 2000*xagr, GRB.MAXIMIZE)

mod.optimize()
```
