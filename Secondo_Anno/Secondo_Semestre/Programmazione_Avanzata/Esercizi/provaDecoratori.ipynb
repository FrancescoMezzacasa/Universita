```python
#voglio decorare la somma
def decoratore(funz):#funz = somma
    def wrapper(*args, **kwargs):#args = (3, 4)
        print(f"Gli addendi sono {args}")
        ris = funz(*args, **kwargs)
        print(f"Ho fatto la somma e vale {ris}")
        return ris
    return wrapper

@decoratore
def somma(a, b):
    return a + b

s = somma(3, 4)
```

```python
#ora voglio farla piu volte quindi usando il decoratore con parametri
def ripetitore(nvolte):
    def decoratore(funz):
        def wrapped(*args, **kwargs):
            print(f"Si vogliono sommare {args} {nvolte} volte")
            ris = 0
            for i in range(nvolte):
                ris += funz(*args, **kwargs)
                print(f"Somma parziale al giro {i + 1} = {ris}")
            print(f"Somma completata, somma totale = {ris}")
            return ris
        return wrapped
    return  decoratore

@ripetitore(10)
def somma(a, b):
    return a + b

s = somma(1, 2)
```
